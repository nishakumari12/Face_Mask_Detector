```python
import os
```

```python
from keras.preprocessing import image
```

```python
import cv2
```

```python
categories = ['with_mask','without_mask']
```

```python
data = []
for category in categories:
    path = os.path.join('train',category)
    
    label = categories.index(category)
    
    for file in os.listdir(path):
        
        img_path = os.path.join(path,file)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(224,224))
        
        data.append([img,label])
```

```python
len(data)
```

```python
import random
```

```python
random.shuffle(data)
```

```python
x = []
y = []

for features,label in data:
    x.append(features)
    y.append(label)
```

```python
len(x)
```

```python
len(y)
```

```python
import numpy as np
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x.shape
```

```python
y.shape
```

```python
x = x/255
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg = VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model = Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
cap = cv2.VideoCapture(2)
```

```python
while True:
    
    ret, frame = cap.read()
    
    #call for detection method
    img = cv2.resize(frame,(224,224))
    
    y_pred = detect_face_mask(img)
    
    coods = detect_face(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    
    for x,y,w,h in coods:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    
    if y_pred == 0:
        draw_label(frame,"No Mask detected",(30,30),(0,255,0))
    else:
        draw_label(frame,"Mask detected",(30,30),(0,0,255))
      
    
    cv2.imshow("window",frame)
    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
        
cv2.destroyAllWindows()
```

```python
def detect_face_mask(img):
    y_pred = model.predict(img.reshape(1,224,224,3))
    
    # Determine the class with the highest probability
    return 0 if y_pred[0][0] > 0.5 else 1
```

```python
def draw_label(img,text,pos,bg_color):
    font = cv2.FONT_HERSHEY_SIMPLEX
    text_size = cv2.getTextSize(text,font,1,cv2.FILLED)
    
    end_x = pos[0] + text_size[0][0] + 2
    end_y = pos[1] + text_size[0][1] - 2
    
    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,font,1,(0,0,0),1,cv2.LINE_AA)
```

```python
haar = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    coods = haar.detectMultiScale(img)
    
    return coods
```
